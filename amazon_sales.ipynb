```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
amazon=pd.read_csv('/content/Copy of Amazon 2_Raw.csv')
amazon
```

```python
table=pd.pivot_table(data=amazon,index='Category',aggfunc=sum)
table
```

```python
fig1=px.line(x=table.index,y=table['Profit'],labels=dict(y='Profit')).show()
```

```python
import calendar
amazon['Order Date'] = pd.to_datetime(amazon['Order Date'])
amazon['Ship Date'] = pd.to_datetime(amazon['Ship Date'])
amazon['Order year'] = amazon['Order Date'].dt.year
amazon['Ship year'] = amazon['Ship Date'].dt.year
amazon
```

```python
import plotly.graph_objects as go

amazon['Ship year']=amazon['Ship year'].astype(str)
amazon['Order year']=amazon['Order year'].astype(str)
ship=pd.pivot_table(amazon,index=['Order year'],aggfunc=sum)
ship
fig3=go.Figure(data=[go.Pie(labels=ship.index,
                           values=ship['Profit'],
                           hole=0.45,
                           pull=[0,0,0,0.2])])
fig3.show()
```

```python
fig4=px.histogram(amazon,x='Order year',y='Sales')
fig4.show()
```
